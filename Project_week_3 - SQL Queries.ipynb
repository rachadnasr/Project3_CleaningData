```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql.cursors
import os
import getpass
```

```python
pw=getpass.getpass()
```

```python
connection_string = "mysql+pymysql://root:"+pw+"@localhost:3306/"
engine=create_engine(connection_string)
```

```python
with engine.connect() as conn : 
    conn.execute(f"CREATE DATABASE IF NOT EXISTS Project_week_3")
```

```python
new_market_data = pd.read_csv("/Users/clemence/Downloads/NewMarkets_Python.csv")
```

```python
new_market_data.to_sql("new_market",engine, "Project_week_3", if_exists="replace", index=True)
```

#### What is the gender repartition for each profession?

```python
query1="""

SELECT Gender, Profession, count(ID)
FROM Project_week_3.new_market
GROUP BY Gender, Profession
ORDER BY Profession, Gender

"""
```

```python
gender_profession=pd.read_sql_query(query1,engine)
gender_profession
```

#### What is the gender repartition depending on the Spending_Score ? 

```python
query2= """

SELECT Gender, Spending_Score, count(ID)
FROM Project_week_3.new_market
WHERE Spending_Score <> "Average"
GROUP BY Gender, Spending_Score
ORDER BY Gender

"""
```

```python
gender_spendings=pd.read_sql_query(query2,engine)
gender_spendings
```

#### What is the profession repartition in the extreme Spending_Score groups ? 

```python
query3="""

select Profession, Spending_Score, count(ID) from Project_week_3.new_market
where Spending_Score <> "Average"
group by Profession, Spending_Score order by Profession, Spending_Score desc;

"""
```

```python
profession_spendings=pd.read_sql_query(query3,engine)
profession_spendings
```

#### What is the average age according to the marital status ? 

```python
query4="""

select new_market.Ever_Married, avg(Age) as Mean_Age from Project_week_3.new_market
group by new_market.Ever_Married;

"""
```

```python
marriage_age=pd.read_sql_query(query4,engine)
marriage_age
```

#### What is tha average age according to the Spending_Score and the Gender ? 

```python
query5="""

select avg(Age) as Avg_Age, Spending_Score, Gender from Project_week_3.new_market
group by Spending_Score, Gender order by Avg_Age desc;

"""
```

```python
spending_age_gender=pd.read_sql_query(query5,engine)
spending_age_gender
```
